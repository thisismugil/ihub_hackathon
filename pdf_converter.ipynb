```python
import os
import google.generativeai as genai
import PyPDF2  # For extracting text from PDFs

# Configure the Generative AI model
genai.configure(api_key='YAIzaSyAxpKqbnQj4PIoEAxNjNiCVUx3hjmPnIWI')  # Replace with your actual API key

generation_config = {
    "temperature": 1.55,
    "top_p": 0.95,
    "top_k": 64,
    "max_output_tokens": 8192,
    "response_mime_type": "text/plain",
}

model = genai.GenerativeModel(
    model_name="gemini-1.5-pro",
    generation_config=generation_config,
    system_instruction="You are a helpful assistant that answers questions based on the extracted PDF content."
)

def extract_text_from_pdf(pdf_path):
    """Extract text from a PDF file."""
    try:
        with open(pdf_path, 'rb') as file:
            reader = PyPDF2.PdfReader(file)
            text = ""
            for page_num in range(len(reader.pages)):
                page = reader.pages[page_num]
                text += page.extract_text() or ""  # Ensure it appends even if text extraction fails
        return text
    except Exception as e:
        print(f"Error reading PDF file: {e}")
        return None

print("Hello! You can provide a PDF file path to extract its content or type 'bye' to exit.")

while True:
    user_input = input("You: ")

    if user_input.lower() == "bye":
        print("Thank you! See you soon.")
        break

    # Check if the user provided a valid PDF file path
    if os.path.isfile(user_input) and user_input.endswith(".pdf"):
        print("Extracting text from the PDF file...")
        pdf_text = extract_text_from_pdf(user_input)

        if pdf_text:
            print("PDF text extracted successfully. You can now ask questions about the content...\n")
            print('-' * 100)
            print(pdf_text)
            print('-' * 100)

            # Allow the user to ask questions about the extracted text
            while True:
                question = input("What would you like to know about the PDF content? (type 'exit' to stop asking questions) ")

                if question.lower() == 'exit':
                    break

                try:
                    # Interact with the Generative AI model
                    response = model.generate(prompt=f"{pdf_text}\n\n{question}")
                    print("AI Response:", response.text)
                except Exception as e:
                    print(f"Error during AI processing: {e}")
        else:
            print("Unable to extract text from the PDF. Try another file.")
    else:
        print("Please provide a valid PDF file path or type 'bye' to exit.")
```
